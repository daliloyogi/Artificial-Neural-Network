# **Preprocessing**

```python
# Scaling features using MinMaxScaler and splitting the dataset
scale = MinMaxScaler()
X = scale.fit_transform(X)
y = scale.fit_transform(y.values.reshape(-1, 1))
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
```

`scale = MinMaxScaler()`:

An instance of MinMaxScaler is created. This scaler is commonly used for scaling numerical features to a specific range, often between 0 and 1.

`X = scale.fit_transform(X)`:

The fit_transform method of the scaler is applied to the features (input variables) X. It scales the numerical features in X to the specified range, transforming them into a new scaled version.

`y = scale.fit_transform(y.values.reshape(-1, 1))`:

Similarly, the target variable y is transformed using the scaler. Since MinMaxScaler expects a 2D array as input, the reshape(-1, 1) method is used to convert the 1D array y into a 2D array before scaling.

`X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)`:

The dataset, consisting of the scaled features (X) and target variable (y), is split into training and testing sets using train_test_split with 80% of the data used for training and 20% for testing.

# **Linear Regression**

```python
#Trying linear regression
lreg = LinearRegression()
lreg.fit(X_train, y_train)
print('Train score', lreg.score(X_train, y_train))
print('Test score', lreg.score(X_test, y_test))
```

The linear regression model is trained using the `LinearRegression` class from scikit-learn. The model is fitted to the scaled training data (`X_train` and `y_train`). Subsequently, the code prints the R-squared scores, which indicate how well the model fits the training and testing datasets. R-squared ranges from 0 to 1, with higher values indicating a better fit. The training score represents the model's performance on the data it was trained on, while the testing score evaluates its ability to generalize to new, unseen data. Comparing these scores helps assess the model's effectiveness and potential for overfitting.

```python
# Making predictions using the trained Linear Regression model on the test set features
y_pred = lreg.predict(X_test)
```

I use the trained linear regression model (`lreg`) to make predictions on the test set features (`X_test`). The `predict` method takes the test features as input and produces corresponding predictions for the target variable. The resulting predictions are stored in the variable `y_pred`. After this step, you can compare these predictions (`y_pred`) with the actual target values (`y_test`) to assess the model's performance on the unseen test data.

# **ANN Model**

```python
# Creating and training an Artificial Neural Network (ANN)
ann = Sequential([Dense(32, activation="relu"),
                 Dense(1, activation='linear')])

ann.compile(optimizer='adam',
            loss='mean_squared_error',
            metrics=['mean_absolute_error'])


history = ann.fit(X_train, y_train, epochs=50, validation_split=0.2)
```

---
In this code , I'm creating and training an artificial neural network (ANN) using the Keras library with TensorFlow backend:

1. `ann = Sequential([Dense(32, activation="relu"), Dense(1, activation='linear')])`:
   - A sequential model (`ann`) is defined, consisting of two layers. The first layer has 32 neurons and uses the ReLU activation function, while the second layer has a single neuron with a linear activation function.

2. `ann.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_absolute_error'])`:
   - The model is configured for training using the `compile` method. The 'adam' optimizer is chosen, 'mean_squared_error' is set as the loss function (suitable for regression tasks), and 'mean_absolute_error' is chosen as the metric to monitor during training.

3. `history = ann.fit(X_train, y_train, epochs=50, validation_split=0.2)`:
   - The model is trained using the `fit` method. It takes the scaled training features (`X_train`) and target variable (`y_train`) as input and runs for 50 epochs. The training process is monitored on a validation subset (`validation_split=0.2`), allowing you to observe the model's performance on a portion of the training data not used for training. The training history is stored in the variable `history`, which can be used for plotting learning curves and evaluating model performance over epochs.

```python
# Extracting training and validation metrics from the training history
tr_acc = history.history['mean_absolute_error']
tr_loss = history.history['loss']
val_acc = history.history['val_mean_absolute_error']
val_loss = history.history['val_loss']

# Finding the epoch with the lowest validation loss and highest validation accuracy
index_loss = np.argmin(val_loss)
val_lowest = val_loss[index_loss]
index_acc = np.argmin(val_acc)
acc_highest = val_acc[index_acc]

# Generating a list of epochs
Epochs = [i+1 for i in range(len(tr_acc))]

# Creating labels for the best epochs
loss_label = f'best epoch= {str(index_loss + 1)}'
acc_label = f'best epoch= {str(index_acc + 1)}'

# Creating a figure with two subplots for training and validation loss, and training and validation accuracy
plt.figure(figsize= (20, 8))
plt.style.use('fivethirtyeight')

# Setting the plot
plt.subplot(1, 2, 1)
plt.plot(Epochs, tr_loss, 'r', label= 'Training loss')
plt.plot(Epochs, val_loss, 'g', label= 'Validation loss')
plt.scatter(index_loss + 1, val_lowest, s= 150, c= 'blue', label= loss_label)
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(Epochs, tr_acc, 'r', label= 'Training Accuracy')
plt.plot(Epochs, val_acc, 'g', label= 'Validation Accuracy')
plt.scatter(index_acc + 1 , acc_highest, s= 150, c= 'blue', label= acc_label)
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Adjusting layout and displaying the figure
plt.tight_layout
plt.show()
```

The provided code will generate a figure with two subplots. Each subplot illustrates the changes in loss and accuracy during the training and validation of an artificial neural network (ANN). The first subplot displays the curves of training loss and validation loss over epochs, marking the epoch with the lowest validation loss. The second subplot shows the curves of training accuracy and validation accuracy, highlighting the epoch with the highest validation accuracy. This visualization is valuable for understanding the model's performance throughout the training and validation processes, pinpointing the epochs with the best loss and accuracy on the validation dataset.
